```python
#equal frequency 
def equil_freq_bin(arr1, m): 
      
    a = len(arr1) 
    n = int(a / m) 
    for i in range(0, m): 
        arr = [] 
        for j in range(i * n, (i + 1) * n): 
            if j >= a: 
                break
            arr = arr + [arr1[j]] 
        print(arr)
```

```python
#equal width 
def equil_width_bin(arr1, m): 
    a = len(arr1) 
    w = int((max(arr1) - min(arr1)) / m) 
    min1 = min(arr1) 
    arr = [] 
    for i in range(0, m + 1): 
        arr = arr + [min1 + w * i] 
    arri=[] 
      
    for i in range(0, m): 
        temp = [] 
        for j in arr1: 
            if j > arr[i] and j < arr[i+1]: 
                temp += [j] 
        arri += [temp] 
    print(arri) 
```

```python
m = 3
  
for a in data:
    print("\n\nequal width binning for column:",a) 
    arr=equil_width_bin(data[a], m) 
```

```python
for a in data:
    print("\n\nEqual Frequency binning for column:",a) 
    arr=equil_freq_bin(data[a], 5) 
```
