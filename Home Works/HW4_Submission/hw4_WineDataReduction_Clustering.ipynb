```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

```

```python
filename = 'C:/Users/subar/Downloads/CMPE-255 Sec 99 - Data Mining/Home Works/HW4_Submission/wine.csv'
data = pd.read_csv(filename)
data.head()
```

```python
# distributing the dataset into two components X and Y (target)
X = data.iloc[:, 1:13].values
y = data.iloc[:, 0].values # target column = class
```

```python
#scatter plot using two columns
data.plot.scatter(x = 'Alcohol', y = 'OD280/OD315of_diluted_wines', c= 'Class', figsize=(12,8), colormap='jet')
```

```python
#centroid calculation
kmeans = KMeans(n_clusters=3, init = 'random', max_iter = 1, random_state = 5).fit(data.iloc[:,[12,1]])
centroids = pd.DataFrame(kmeans.cluster_centers_, columns = list(data.iloc[:,[12,1]].columns.values))

fig, ax = plt.subplots(1, 1)
data.plot.scatter(x = 'Alcohol', y = 'OD280/OD315of_diluted_wines', c= kmeans.labels_, figsize=(12,8), colormap='jet', ax=ax, mark_right=False)
centroids.plot.scatter(x = 'Alcohol', y = 'OD280/OD315of_diluted_wines', ax = ax,  s = 80, mark_right=False)
```

```python
#recalculating centroid by taking mean of the clusters for better result

kmeans = KMeans(n_clusters=3, init = 'random', max_iter = 100, random_state = 5).fit(data.iloc[:,[12,1]])
centroids = pd.DataFrame(kmeans.cluster_centers_, columns = list(data.iloc[:,[12,1]].columns.values))

fig, ax = plt.subplots(1, 1)
data.plot.scatter(x = 'Alcohol', y = 'OD280/OD315of_diluted_wines', c= kmeans.labels_, figsize=(12,8), colormap='jet', ax=ax, mark_right=False)
centroids.plot.scatter(x = 'Alcohol', y = 'OD280/OD315of_diluted_wines', ax = ax,  s = 80, mark_right=False)
```
